Problema 3:

 Un número al que se le resta 7 da como resultado 18.Planteamiento: Definimos nuestra incógnita como $x$ (el número desconocido). La frase "se le resta 7" se traduce como $- 7$ y "da como resultado" como $= 18$.Ecuación: $x - 7 = 18$Despeje: Para dejar a la $x$ sola, el 7 que está restando pasa al otro lado sumando.$x = 18 + 7$Solución: $x = 25$

```python
# Problema 3: x - 7 = 18
# resultado = 18
# resta = 7
resultado_total = int(input("Ingrese el resultado total: "))
resta_total = int(input("Ingrese la resta"))
resta = 7

# Despejando: x = resultado + resta
numero = resultado_total + resta

print(f"El número buscado en el problema 3 es: {numero}")
```

Problema 4:


 El triple de un número más 4 es igual a 31.Planteamiento: "El triple de un número" es $3x$. "Más 4" es $+ 4$. "Igual a 31" es $= 31$.Ecuación: $3x + 4 = 31$Despeje:Primero, el 4 que suma pasa restando: $3x = 31 - 4$ $\rightarrow$ $3x = 27$.Luego, el 3 que multiplica a la $x$ pasa dividiendo: $x = \frac{27}{3}$.Solución: $x = 9$

```python
# Problema 4: 3x + 4 = 31
# total = 31
# suma = 4
# multiplo = 3
total = int(input("Ingrese el total: "))
suma = int(input("Ingrese la suma: "))
multiplo = 3

# Despejando: x = (total - suma) / multiplo
x = (total - suma) / multiplo

print(f"El número buscado en el problema 4 es: {int(x)}")
```

Problema 5:


 Terreno con largo 5m más que el ancho y perímetro de 50m.Planteamiento:Ancho $= x$Largo $= x + 5$El perímetro de un rectángulo es $2 \cdot \text{largo} + 2 \cdot \text{ancho} = \text{Perímetro}$.Ecuación: $2(x + 5) + 2(x) = 50$Despeje:Distribuimos el 2: $2x + 10 + 2x = 50$.Simplificamos términos semejantes: $4x + 10 = 50$.Restamos 10: $4x = 40$.Dividimos entre 4: $x = \frac{40}{4} = 10$.Solución: * Ancho $= 10\text{m}$Largo $= 10 + 5 = 15\text{m}$

```python
# Problema 5: 2(x + 5) + 2x = 50 -> 4x + 10 = 50
# perimetro = 50
# diferencia_largo = 5 # El largo es x + 5
# La ecuación simplificada es: 4x + (2 * diferencia_largo) = perimetro
perimetro = int(input("Ingrese el perímetro: "))
diferencia_largo = int(input("Ingrese la diferencia de largo: "))

# Despejando el ancho (x)
ancho = (perimetro - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"Las medidas del terreno son: Ancho = {ancho}m, Largo = {largo}m")
```
